Clean Data:  The column "Code" was confusing so it was renamed to abbrevation as to make it more simpler to understand since they are abreviations of he Country/Territory. All duplicates are dropped if there are any. The data is now cleaned up. To make sure the data was displayed.

```python
# rename columns and get rid of dduplicate information
d.rename(columns={'Code': 'Abbreviation'}, inplace=True)  
d = d.drop_duplicates()
print(d)
```

Visualization: Below is the code to plotting the data and making it visual in order to understand in more detail which diseases are more deadly between the years 1990 and 2019 across the globe. Made a knew variable (disease_columns) to hold every column that is a disease. Added all the deaths in each column and stored into total_deaths. Set the plot size and made a palette to make it easier to understand the graph and differentiate between each bar. Plotted the bar with the x axis as disease and y as the deaths. Gave a title and label and roated the disease names so they no longer overlap.

```python
# Extract disease columns (start after 3rd col)
disease_columns = d.columns[3:]

# Sum deaths across countries & years
total_deaths = d[disease_columns].sum()

# Create DataFrame for plotting
deaths = pd.DataFrame({'Disease': disease_columns, 'Total Deaths': total_deaths})

# Plot
plt.figure(figsize=(15, 8))
palette = sns.color_palette("husl", len(disease_columns))
sns.barplot(x='Disease', y='Total Deaths', data=deaths, palette=palette)

plt.title('Global Deaths by Disease (1990–2019)', fontsize=16)
plt.xlabel('Disease', fontsize=12)
plt.ylabel('Total Deaths', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Dropped the 2 largest disaeases since they had such a greater amount of deaths. Will regraph the the disaese so its easier to compare the diseases with less deaths.

```python
d= d.drop(columns = ['Cardiovascular Diseases', 'Neoplasms'])
```

Remake the chart without the two largest causes of death to compare the lower causes of deaths.

```python
#The diseases start after the third column
disease_columns = d.columns[3:]
total_deaths = d[disease_columns].sum() #sum all the deaths in each column

# Create a new DataFrame for plotting on graph
deaths = pd.DataFrame({'Disease': disease_columns, 'Total Deaths': total_deaths})
#size of the graph
plt.figure(figsize=(15, 10))

# color palette for each disease to make it easier to differentiate
palette = sns.color_palette("husl", len(disease_columns))

# Use a bar plot (sns.barplot) with the new DataFrame
sns.barplot(x='Disease', y='Total Deaths', data=deaths, palette=palette)

#Title and x,y label
plt.title('Deaths caused by Diseases')
plt.xlabel('Disease')
plt.ylabel('Deaths')

# Rotate label for better readability
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```
